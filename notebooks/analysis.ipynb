```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# global variables
ROOT_URL = '../data/facebook/'
TRAIN_CSV = 'train.json'
TEST_CSV = 'test.json'
VAL_CSV = 'dev.json'
```

```python
# read dataframes
train_df = pd.read_json(os.path.join(ROOT_URL, TRAIN_CSV))
test_df = pd.read_json(os.path.join(ROOT_URL, TEST_CSV))
val_df = pd.read_json(os.path.join(ROOT_URL, VAL_CSV))
```

```python
train_arr = []
test_arr = []
val_arr = []

train_arr.append(len(train_df[train_df['label'] == 0]) * 100 / len(train_df))
train_arr.append(len(train_df[train_df['label'] == 1]) * 100 / len(train_df))

test_arr.append(len(test_df[test_df['label'] == 0]) * 100 / len(test_df))
test_arr.append(len(test_df[test_df['label'] == 1]) * 100 / len(test_df))

val_arr.append(len(val_df[val_df['label'] == 0]) * 100 / len(val_df))
val_arr.append(len(val_df[val_df['label'] == 1]) * 100 / len(val_df))
```

```python
plt.subplot(2, 2, 1)
plt.title("Train")
plt.pie(train_arr, labels = ['Non_Hateful', 'Hateful'], autopct='%1.1f%%')

plt.subplot(2, 2, 2)
plt.title("Test")
plt.pie(test_arr, labels = ['Non Hateful', 'Hateful'], autopct='%1.1f%%')

plt.subplot(2, 2, 3)
plt.title("Validation")
plt.pie(val_arr, labels = ['Non Hateful', 'Hateful'], autopct='%1.1f%%')

plt.show()
```
