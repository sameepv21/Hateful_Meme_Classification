```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# global variables
ROOT_URL = '../data/facebook/'
TRAIN_JSON = 'train.json'
DEV_JSON = 'dev.json'
```

```python
# read dataframes
train_df = pd.read_json(os.path.join(ROOT_URL, TRAIN_JSON))
dev_df = pd.read_json(os.path.join(ROOT_URL, DEV_JSON))
```

```python
train_arr = []
dev_arr = []

train_arr.append(len(train_df[train_df['label'] == 0]) * 100 / len(train_df))
train_arr.append(len(train_df[train_df['label'] == 1]) * 100 / len(train_df))

dev_arr.append(len(dev_df[dev_df['label'] == 0]) * 100 / len(dev_df))
dev_arr.append(len(dev_df[dev_df['label'] == 1]) * 100 / len(dev_df))
```

```python
plt.subplot(2, 2, 1)
plt.title("Train")
plt.pie(train_arr, labels = ['Non_Hateful', 'Hateful'], autopct='%1.1f%%')

plt.subplot(2, 2, 2)
plt.title("Validation")
plt.pie(dev_arr, labels = ['Non Hateful', 'Hateful'], autopct='%1.1f%%')

plt.show()
```
