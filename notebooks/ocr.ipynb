```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/facebook/dev/hateful/01456.png')
img = cv2.bilateralFilter(img, 10, 50,60)
_, img = cv2.threshold(img, 240, 255, 1) # Image thresholding
```

```python
plt.imshow(img, cmap='gray')
plt.show()
```

```python
custom_config = r'--oem 3 --psm 11'
text = pytesseract.image_to_string(img, config=custom_config)
print(text.replace('\n', ' '))
```
